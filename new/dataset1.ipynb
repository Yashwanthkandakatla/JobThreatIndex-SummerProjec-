```python
percentile25=df3['Tasks'].quantile(0.25)
percentile75=df3['Tasks'].quantile(0.75)
iqr=percentile75-percentile25
upperlimitpm1=percentile75+1.5*iqr
lowerlimitpm1=percentile25-1.5*iqr
df3['Tasks']=np.where(df3['Tasks']>upperlimitpm1,upperlimitpm1,np.where(df3['Tasks']<lowerlimitpm1,lowerlimitpm1,df3['Tasks']))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df3.drop('AI Impact',axis=1), 
                                                    df3['AI Impact'], test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=12)
```

```python
knn.fit(X_train.values,y_train.values)
```

```python
y_pred = knn.predict(X_test.values)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

```python
from sklearn.metrics import r2_score

r2=r2_score(y_test,y_pred)
print(r2)
```

```python
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(mape)
```

```python
from sklearn.metrics import explained_variance_score
explained_var = explained_variance_score(y_test, y_pred)
print(explained_var)
```
